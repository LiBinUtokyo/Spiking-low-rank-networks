```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_total = 1000
Smn = 0
Sm = 2
Sn = 2
factor_mn  = 0.01
ones = np.ones(N_total)
m_1, n_1 = np.random.multivariate_normal([0,0],[[Sm**2, Smn],[Smn, Sn**2]],N_total).transpose()
m_1 = m_1 - (np.dot(m_1,ones)/np.dot(ones,ones))*ones
n_1 = n_1 - (np.dot(n_1,ones)/np.dot(ones,ones))*ones

rank_1 = factor_mn * np.outer(m_1, n_1)/N_total
#绘制矩阵
plt.figure()
plt.imshow(rank_1, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Rank 1 matrix')
plt.show()
```

```python

N_total = 1000
C = 100
N_exc = int(0.8 * N_total)
N_inh = int(0.2 * N_total)
f = C/N_total
# f = 0.1 # sparsity
C_e = int(f * N_exc)
C_i = int(f * N_inh)
J_value = 0.6
g = 5


E_I = np.zeros((N_total, N_total))
for i in range(0, N_total):
    ind_exc = random.sample(range(0, N_exc), C_e)
    ind_inh = random.sample(range(N_exc, N_inh+N_exc), C_i)
    E_I[i, ind_exc] = J_value
    E_I[i, ind_inh] = -g*J_value
m_2 = np.ones(N_total)
n_2 = np.zeros(N_total)
n_2[:N_exc] = J_value*C_e/N_exc
n_2[N_exc:] = -g*J_value*C_i/N_inh

rank = np.linalg.matrix_rank(E_I)
print("矩阵的秩:", rank)

# 画出权重矩阵
plt.figure()
plt.imshow(E_I, cmap='hot', interpolation='nearest')
plt.colorbar()
plt.title('Weight Matrix')
plt.show()
plt.figure()
plt.imshow(np.outer(m_2, n_2), cmap='hot', interpolation='nearest')
plt.colorbar()
plt.show()

# 检查非零元素的比例
print("非零元素的比例:", np.count_nonzero(E_I) / (N_total * N_total))
```
